### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Letâ€™s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Select and clean relevant column
df_databases = df[['DatabaseWantToWorkWith']].dropna()

# Step 2: Split multiple selections into individual rows
df_databases['DatabaseWantToWorkWith'] = df_databases['DatabaseWantToWorkWith'].str.split(';')
df_exploded = df_databases.explode('DatabaseWantToWorkWith')

# Step 3: Clean whitespace
df_exploded['DatabaseWantToWorkWith'] = df_exploded['DatabaseWantToWorkWith'].str.strip()

# Step 4: Count frequency of each database
db_counts = df_exploded['DatabaseWantToWorkWith'].value_counts().nlargest(5)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(db_counts, labels=db_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.tight_layout()
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Letâ€™s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Select the column and drop missing values
df_os = df[['OpSysProfessional use']].dropna()

# Step 2: Split multiple selections into individual rows
df_os['OpSysProfessional use'] = df_os['OpSysProfessional use'].str.split(';')
df_exploded = df_os.explode('OpSysProfessional use')

# Step 3: Clean whitespace
df_exploded['OpSysProfessional use'] = df_exploded['OpSysProfessional use'].str.strip()

# Step 4: Count operating system frequency
os_counts = df_exploded['OpSysProfessional use'].value_counts().nlargest(6)

# Step 5: Plot pie chart
plt.figure(figsize=(8, 6))
plt.pie(os_counts, labels=os_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Top Operating Systems Used Professionally')
plt.axis('equal')  # Keeps pie chart circular
plt.tight_layout()
plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
##Write your code here

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Select and clean the relevant column
df_web = df[['WebframeWantToWorkWith']].dropna()

# Step 2: Split semicolon-separated values into individual rows
df_web['WebframeWantToWorkWith'] = df_web['WebframeWantToWorkWith'].str.split(';')
df_exploded = df_web.explode('WebframeWantToWorkWith')

# Step 3: Remove leading/trailing whitespace
df_exploded['WebframeWantToWorkWith'] = df_exploded['WebframeWantToWorkWith'].str.strip()

# Step 4: Count the top 5 most wanted web frameworks
web_counts = df_exploded['WebframeWantToWorkWith'].value_counts().nlargest(5)

# Step 5: Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    web_counts,
    labels=web_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Pastel2.colors
)
plt.title('Top 5 Preferred Web Frameworks')
plt.axis('equal')  # Draw pie as a circle
plt.tight_layout()
plt.show()
```
